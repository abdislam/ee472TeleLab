## Introduction
In computers, we cannot have a continuous signal since it means it is defined at every point, infinite number of points require unlimited memory! But instead, we will define the signal at different time intervals and this will convert the signal from continuous to discrete. The signal value at a time interval is called a sample and the number of samples in one second is called the sampling frequency. The sampling process will limit the bandwidth of the signal to the half of sampling frequency, Nyquist theorem.


````octave
% Define the sampling frequency
Fs = 1e2;

% Determine the duration of the signal in seconds
du = 1; % one second

% Create the time vector
t  = 0: 1/Fs: du;

% Create the frequency vector
f  = -Fs/2: 1/du :Fs/2;

% Specify the frequency of a signal
fc = 10; % note that fc, in Hz, should be less than Fs/2

% Create the signal time vector
y  = cos(2*pi*fc*t);

% Convert the signals to frequency domain
Y  = abs( fftshift( fft(y)/length(y) ));

% Plot the signal in each domain
figure, plot(t,y); title('time domain')
figure, plot(f,Y); title('frequency domain')
````

The previous steps are very helpful as a starter. Yet they are very inefficient and difficult to make corrections in longer programs.
In MATLAB, you can define a function for repeated operations so you do not have to write them more and more.
Define a function for converting signals from time to frequency domain:

````octave
function X = sig_spec(x)
% sig_spec, is just a name, you can use your own, determines the spectrum of x by calculating FFT
% X is the signal in frequency domain
% x is the signal
X = abs( fftshift( fft(x)/length(x) ));

end

% test your function
figure, plot(f,sig_spec(y));title('frequency domain')
````
