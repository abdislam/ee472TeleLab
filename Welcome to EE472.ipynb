## Introduction
In computers, we cannot have a continuous signal since it means it is defined at every point, infinite number of points require unlimited memory! But instead, we will define the signal at different time intervals and this will convert the signal from continuous to discrete. The signal value at a time interval is called a sample and the number of samples in one second is called the sampling frequency. The sampling process will limit the bandwidth of the signal to the half of sampling frequency, Nyquist theorem.

•	Define the sampling frequency: ```Fs = 10000; % Hz ```

•	Determine the duration of the signal in seconds: ```du = 1; % sec.```

•	Create the time vector: ```t = 0: 1/Fs: du;```

•	Create the frequency vector: ```f = -Fs/2: 1/du :Fs/2; ```

•	Specify the frequency of a signal: ```fc = 1000;  note that fc, in Hz, should be less than Fs/2 ```

•	Create the signal time vector: ```y = cos(2*pi*fc*t);```

•	Convert the signals to frequency domain: ```Y = abs( fftshift( fft(y)));```

•	Plot the signal in each domain: ```figure, plot(t,y); figure, plot(f,Y);```

The previous steps are very helpful as a starter. Yet they are very inefficient and difficult to make corrections in longer programs. In MATLAB, you can define a function for repeated operations so you do not have to write them more and more.


- Now, you are asked to create two sinusoidal signals of 10 Hz and 100 Hz frequencies and find: First, the multiplication, use element multiplication. Second, the summation. Third, the exponent of the summation minus 1. Compare the three outputs in time and frequency domain and comment.

````octave
clear all
Fs = 1000;
du = 1;
t  = 0: 1/Fs: du;
f  = -Fs/2: 1/du: Fs/2;
fc = 100;
fm = 10;
c  = cos(2*pi*fc*t);
m  = cos(2*pi*fm*t);
y1 = c + m;
y2 = exp(y1) - 1;
y3 = c .* m + c;
Y1 = abs(fftshift(fft(y1)));
Y2 = abs(fftshift(fft(y2)));
Y3 = abs(fftshift(fft(y3)));

figure;
subplot(2,3,1), plot(t,y1);
subplot(2,3,2), plot(t,y2);
subplot(2,3,3), plot(t,y3);
subplot(2,3,4), plot(f,Y1);
subplot(2,3,5), plot(f,Y2);
subplot(2,3,6), plot(f,Y3);
````
